```r
path <- "/content/16S_project"  

fnFs <- sort(list.files(path, pattern = "\\.fastq(.gz)?$", full.names = TRUE))
cat("Files found:", length(fnFs), "\n")
print(basename(fnFs))
sample.names <- sub("\\.fastq(.gz)?$", "", basename(fnFs))

# Create filtered folder if missing
filt_path <- file.path(path, "filtered2")
if(!dir.exists(filt_path)) dir.create(filt_path)

filtFs <- file.path(filt_path, paste0(sample.names, "_filt.fastq.gz"))

# Run filtering and trimming
out <- filterAndTrim(fnFs, filtFs,
                     truncLen=150,
                     maxN=0,
                     maxEE=2,
                     truncQ=2,
                     rm.phix=TRUE,
                     compress=TRUE,
                     multithread=TRUE)

print(out)
```

```r
library(dada2)
library(phyloseq)
library(ggplot2)

# Set paths
path <- "/content/16S_project/filtered2"
filtFs <- sort(list.files(path, pattern = "_filt.fastq.gz$", full.names = TRUE))
sample.names <- sapply(strsplit(basename(filtFs), "_filt"), `[`, 1)

#Learn error rates
if (file.exists("errF.rds")) {
  errF <- readRDS("errF.rds")
  message("Loaded saved error model.")
} else {
  errF <- learnErrors(filtFs, multithread = TRUE)
  saveRDS(errF, "errF.rds")
}

#Dereplicate
if (file.exists("derepFs.rds")) {
  derepFs <- readRDS("derepFs.rds")
  message("Loaded saved dereplication.")
} else {
  derepFs <- derepFastq(filtFs, verbose = TRUE)
  names(derepFs) <- sample.names
  saveRDS(derepFs, "derepFs.rds")
}

#Sample inference
if (file.exists("dadaFs.rds")) {
  dadaFs <- readRDS("dadaFs.rds")
  message("Loaded saved dada inference.")
} else {
  dadaFs <- dada(derepFs, err = errF, multithread = TRUE)
  saveRDS(dadaFs, "dadaFs.rds")
}

#Sequence table
if (file.exists("seqtab.rds")) {
  seqtab <- readRDS("seqtab.rds")
} else {
  seqtab <- makeSequenceTable(dadaFs)
  saveRDS(seqtab, "seqtab.rds")
}

#Remove chimeras
if (file.exists("seqtab_nochim.rds")) {
  seqtab.nochim <- readRDS("seqtab_nochim.rds")
} else {
  seqtab.nochim <- removeBimeraDenovo(seqtab, method="consensus", multithread=TRUE, verbose=TRUE)
  saveRDS(seqtab.nochim, "seqtab_nochim.rds")
}

cat("Reads retained after chimera removal:", round(sum(seqtab.nochim) / sum(seqtab) * 100, 2), "%\n")



```

```r
library(phyloseq)

if (file.exists("phyloseq_object.rds")) {
  ps <- readRDS("phyloseq_object.rds")
  message("Loaded saved phyloseq object.")
} else {
  ps <- phyloseq(
    otu_table(seqtab.nochim, taxa_are_rows = FALSE),
    tax_table(taxa)
  )
  saveRDS(ps, "phyloseq_object.rds")
  message("Created and saved phyloseq object.")
}
```

```r
library(dplyr)
library(ggplot2)

taxa_df <- as.data.frame(taxa)

genus_counts <- taxa_df %>%
  group_by(Genus) %>%
  summarise(num_ASVs = n()) %>%
  arrange(desc(num_ASVs))

genus_counts_top20 <- genus_counts[1:20, ]

ggplot(genus_counts_top20, aes(x = reorder(Genus, -num_ASVs), y = num_ASVs)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  labs(title = "Number of ASVs per Genus",
       x = "Genus",
       y = "Number of ASVs")
```

```r
library(phyloseq)

ps_phylum <- tax_glom(ps, taxrank = "Phylum")  # Aggregate counts by Phylum
ps_phylum_rel <- transform_sample_counts(ps_phylum, function(x) x / sum(x))  # Convert to relative abundance

plot_bar(ps_phylum_rel, fill = "Phylum") +
  theme_bw() +
  ggtitle("Relative Abundance at Phylum Level")
```
